```python
from lxml import html
import requests

page = requests.get('http://www.imdb.com/chart/top')
tree = html.fromstring(page.content)

links = tree.xpath('//td[@class="titleColumn"]/a')

for link in links:
    movie_title = link.text
    cast = link.attrib['title']
    detail_url = 'http://www.imdb.com/' + link.attrib['href']
    # open detail page
    detail_page = requests.get(detail_url)
    detail_tree = html.fromstring(detail_page.content)
    description = detail_tree.xpath('//p[@itemprop="description"]/text()')
    print "TITLE " + movie_title
    print "CAST " + cast
    print "DETAIL URL " + detail_url
    print "DESCRIPTION " + "".join(description)
    print ""
```
